```python
office_supplies.groupby(['Region', 'Category'])['Quantity'].sum().plot(kind='pie', subplots=True,startangle=90,figsize=(15,10), 
                                        autopct='%1.1f%%')
```

We can see from the pie chart that for all the regions the category that sold the lowest quantity of products was the technology category (i.e. Central -> 4.1%, West -> 6.2%, South -> 3.0%, East -> 5.1%).

It can also be observed that the Office Supplies Category sold the most products in all 4 regions.

```python
technology = office_supplies[office_supplies['Category'] == 'Technology']
```

```python
technology.groupby(['Region', 'Sub-Category'])['Quantity'].sum().plot(kind='pie', subplots=True, startangle=90,figsize=(20,15), 
                                        autopct='%1.1f%%')
```

From the pie chart, it becomes obvious that out of the products sold that belong to the technology category; the products in the sub-category of Machines and Copiers seem to be the least ordered in all 4 regions. Even though, the quantities are low, they seem to be ordered more in East and West a little more than the other regions.

The sub-category of phones seems to be the best perfoming sub-category in 3 out of 4 regions. It is only slightly outperformed in the Central region by 0.1%.

It can also be observed that the percentage of phones and accesories ordered are almost identical in all 4 regions. This may mean that the people who ordered phones also ordered phones on the same day or at a later day.

```python
# Converting Order Date column to a numeric column
office_supplies['Order Date'] = pd.to_datetime(office_supplies['Order Date'])
```

```python
office_supplies
```

```python
office_supplies.groupby(office_supplies['Order Date'].dt.year)['Sales'].sum().keys()
```

```python
office_supplies.groupby(office_supplies['Order Date'].dt.year)['Sales'].count()
```

2014 has the lowest number of entries & 2017 has the highest. 

```python
# Plotting YEAR against Sales Sum in each Year
years = ['2014', '2015', '2016', '2017']
plt.plot(years, office_supplies.groupby(office_supplies['Order Date'].dt.year)['Sales'].sum())
plt.show()
```

From the Order Dates of the data, we can see that sales dropped from 2014 to 2015, but drastically increased in 2016 and 2017.
